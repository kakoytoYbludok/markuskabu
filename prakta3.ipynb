Класс "OrdinalEncoder" предоставляет возможность преобразования категориальных признаков в целочисленные значения, сохраняя при этом порядок. Каждому уникальному значению категориального признака присваивается уникальное целое число, и значения рассматриваются в упорядоченном порядке по возрастанию.Импортируем этот класс.


```python
from sklearn.preprocessing import OrdinalEncoder
```

Код OEnc = OrdinalEncoder() создает экземпляр класса OrdinalEncoder и присваивает его переменной OEnc.



Затем, метод fit() вызывается на объекте OEnc с аргументом df[["Genre"]]. Это позволяет кодировщику OrdinalEncoder "обучиться" на уникальных значениях столбца "Genre" из датафрейма df.



```python
# Создание экземпляра класса OrdinalEncoder
OEnc = OrdinalEncoder()

# Обучение OrdinalEncoder на уникальных значениях столбца "Genre"
OEnc.fit(df[["Genre"]])
```

OEnc.categories_ возвращает массив, содержащий уникальные значения категориального признака, на котором был обучен OrdinalEncoder. Каждый элемент массива соответствует одному столбцу категориальных данных и содержит уникальные значения этого столбца в упорядоченном виде



```python
# Получение уникальных значений категориального признака
categories = OEnc.categories_
```

OEnc.transform(df[["Genre"]]) применяет обученный OrdinalEncoder к указанному столбцу "Genre" в DataFrame df и преобразует его категориальные значения в соответствующие целочисленные значения с сохранением порядка.



```python
# Преобразование категориальных значений столбца "Genre" в целочисленные значения с сохранением порядка
transformed_data = OEnc.transform(df[["Genre"]])
```

```python
df[["Genre"]] = OEnc.transform(df[["Genre"]]) #Перенос преобразованных данных в изначальный датасет
df #Вывод данных из датасета
```

После применения "OrdinalEncoder", мы получаем новый датасет, в котором значения жанра преобразованы в целочисленные значения, сохраняя при этом их порядок.



"LabelEncoder" - это класс библиотеки, который используется для преобразования категориальных данных в числовые значения. В отличие от "LabelEncoder", "OrdinalEncoder" также сохраняет порядок значений в каждом столбце при кодировании категориальных признаков.

"OneHotEncoder" является инструментом, используемым для преобразования категориальных данных в бинарные значения. Сначала импортируется необходимая библиотека. Затем создается вектор Y, который представляет собой целевую переменную. Далее создается матрица X, содержащая признаки (все столбцы, кроме "Global_Sales"). В конечном итоге датасет разделяется на обучающую и тестовую выборки с помощью метода train_test_split. Этот метод принимает матрицу признаков X, вектор целевой переменной Y, размер тестовой выборки test_size (выраженный в долях или числе) и random_state, который задает начальное состояние для случайного разделения данных.



```python
from sklearn.model_selection import train_test_split #Подключение библиотеки
```

В данном коде происходит создание вектора Y, который содержит значения целевой переменной "Global_Sales" из исходного датасета. Затем создается матрица X, которая является исходным датасетом df без столбца "Global_Sales". Наконец, выводится вектор Y, содержащий значения целевой переменной "Global_Sales".



```python
# Создание вектора Y, который представляет собой целевую переменную "Global_Sales"
Y = df["Global_Sales"]

# Создание матрицы X путем удаления столбца "Global_Sales" из исходного датасета
X = df.drop(columns="Global_Sales")

# Вывод вектора Y, содержащего значения целевой переменной "Global_Sales"
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 40) #Разделение данных на четыре массива: X_train, X_test, y_train и y_test, которые представляют собой обучающий и тестовый наборы признаков и целевых переменных соответственно.
print("Кол-во входных обущающих данных: ",len(X_train))
print("Кол-во входных тестовых данных: ",len(X_test)) 
print("Кол-во проверочных обущающих данных: ",len(Y_train)) 
print("Кол-во проверочных тестовых данных: ",len(Y_test)) #Вывод длины данных
```

MinMaxScaler - это класс из библиотеки sklearn.preprocessing, который используется для нормализации данных путем приведения их к определенному диапазону значений. Он преобразует каждый признак независимо, масштабируя значения признаков таким образом, чтобы они находились в диапазоне от 0 до 1.

Сначала импортируется необходимая библиотека. Затем применяется метод fit_transform к столбцам "Other_Sales" и "Global_Sales". Этот метод адаптируется к данным, вычисляет минимальное и максимальное значение каждого признака, а затем нормализует значения признаков, приводя их в диапазон от 0 до 1.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Применение метода fit_transform к столбцам 'Other_Sales' и 'Global_Sales' для нормализации данных
scaled_features = scaler.fit_transform(df[['Other_Sales','Global_Sales']])

# Создание нового датафрейма df_scaled с нормализованными признаками
df_scaled = pd.DataFrame(scaled_features, columns=['Other_Sales','Global_Sales'])

# Вывод нормализованных данных df_scaled
df_scaled
```

Результатом применения нормализации и масштабирования на числовых признаках "Other_Sales" и "Global_Sales" является новый датафрейм df_scaled, в котором эти признаки теперь находятся в диапазоне от 0 до 1. Это позволяет сравнивать значения этих признаков относительно других признаков в датасете и облегчает анализ и интерпретацию данных.



Разделение датасета на обучающую и тестовую выборку с учетом дисбаланса значений у целевой переменной Для того чтобы учесть дисбаланс значений у целевой переменной при разделении датасета на обучающую и тестовую выборку, можно использовать стратифицированное разбиение. Это означает, что соотношение классов в исходном датасете будет сохранено в обеих выборках.



В результате выполнения данного кода мы получаем четыре переменные: X_train, X_test, y_train, y_test, которые представляют собой обучающую и тестовую выборки признаков и целевой переменной соответственно. При разделении датасета мы учитываем дисбаланс значений целевой переменной с помощью параметра stratify, который задает список классов [0, 1] для стратифицированного разбиения. Таким образом, соотношение классов будет сохранено как в обучающей, так и в тестовой выборке. Параметр random_state устанавливает начальное состояние для случайного разбиения данных, чтобы результаты были воспроизводимыми.



```python
# Импортируем необходимую библиотеку для разделения датасета
from sklearn.model_selection import train_test_split

# Задаем матрицу признаков X и вектор целевой переменной y
X = df.drop(columns=['Global_Sales'])

y = df['Global_Sales']

# Выполняем разделение датасета на обучающую и тестовую выборки
# Параметр test_size указывает размер тестовой выборки (в данном случае 20%)
# Параметр stratify=[0, 1] гарантирует стратифицированное разбиение, сохраняя дисбаланс классов
# Параметр random_state=42 задает начальное состояние для случайного разбиения данных
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1, stratify=[0, 0.9], random_state=42)
```

1)OrdinalEncoder: Используется для преобразования категориальных признаков с упорядоченными значениями в числовые, назначая каждому уникальному значению соответствующее целочисленное значение.
